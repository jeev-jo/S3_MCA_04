```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Dataset

```python
df = pd.read_csv('Mall_Customers.csv')
df.describe()
```

## Dataset Preprocessing

### Removing duplicates

```python
df.drop_duplicates(inplace=True)
df.describe()
```

### Extracting Independent variables
selecting only annual income and spending score

```python
x = df.iloc[:,[3,4]].values
print(x)
```

## Fitting data using ELBOW Method

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
print(wcss)
```

## Visualisation

### Lineplot

```python
plt.figure(figsize=(10,5))
sns.lineplot(range(1,11), wcss,marker='o',color='b')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## Fitting K-Means to the dataset

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_pred = kmeans.fit_predict(x)
print(y_pred)
```

## Visualisation of clusters

```python
plt.scatter(x[y_pred == 0, 0], x[y_pred == 0, 1], s = 100, c = 'b', label = 'Cluster 1')
plt.scatter(x[y_pred == 1, 0], x[y_pred == 1, 1], s = 100, c = 'g', label = 'Cluster 2')
plt.scatter(x[y_pred == 2, 0], x[y_pred == 2, 1], s = 100, c = 'cyan', label = 'Cluster 3')
plt.scatter(x[y_pred == 3, 0], x[y_pred == 3, 1], s = 100, c = 'plum', label = 'Cluster 4')
plt.scatter(x[y_pred == 4, 0], x[y_pred == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 200, c = 'r', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(loc=(1.04, 0))
plt.show()
```
