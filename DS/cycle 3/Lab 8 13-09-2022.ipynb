### b)Write a program to draw multivariate visualization plots(scatter plots, scatter multiple,scatter matrix,bubble plot) with matplotlib for iris dataset

### Scatter plot

```python
df.plot.scatter(x='petal_length', y = 'petal_width')
```

### Scatter Multiple

```python
ax1 = df.plot(kind='scatter', x='petal_length', y = 'petal_width', label='petal_length Vs petal_width', color='r')    
ax2 = df.plot(kind='scatter', x='petal_length', y = 'sepal_length', label='petal_length Vs sepal_length',color='g',ax=ax1)    
ax3 = df.plot(kind='scatter',  x='petal_length', y = 'sepal_width',label='petal_length Vs sepal_width', color='b',ax=ax2)
print(ax1 == ax2 == ax3)
```

### Scatter Matrix

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df, alpha=0.5, diagonal='kde')
pd.plotting.scatter_matrix(df) 
```

### Bubble Plot

```python
plt.scatter('petal_length', 'petal_width',  s='sepal_length',  alpha=0.5,  data=df)
```
